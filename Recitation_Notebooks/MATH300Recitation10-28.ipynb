# Exam: Problem 5(b)

Given _any_ $\alpha > 1$, give an example of a sequence $p_{n}$ that converges with that order.

## Solution:

It is important to recognize that YOU are given an arbitrary $\alpha$. Many of you read this as "pick whichever $\alpha$ you'd like", which is not what the question calls for (indeed, if that was what was asked, you could just pick $\alpha = 2$ and reference part (a)). Perhaps a more obvious notation for what we are looking for is a _parametrized family_ $p_{n}(\alpha)$ of sequences. Then, for any $\alpha > 1$, our goal is to show that $p_{n}(\alpha)$ converges with order $\alpha$.

Now that we know _what_ we need to find, how do we find it? Perhaps the most obvious guess (in the light of the analysis from part (a)) is $p_{n}(\alpha) = 10^{-\alpha^{n}}$ for all $\alpha > 1$. Of course, this aligns with our result from (a), as $p_{n}(2)$ converges with order $2$. Now, we just have to prove that for _every_ $\alpha > 1$, we get the appropriate order of convergence. Since $\alpha > 1$, the exponent blows up in the negative direction (i.e., $-\alpha^{n} \to -\infty$ as $n \to \infty$), so it should be clear that the limit of every sequence $p_{n}(\alpha)$ is zero. Then, we use the definition given in the beginning of Section 2.4, which says that if $\alpha', \lambda > 0$ (using $\alpha'$ to remove ambiguity about known and unknown quantities. We know $\alpha$, and want to show $\alpha' = \alpha$) exist with $\newline \newline$
$$
\lim_{n \to \infty}\frac{|p_{n + 1} - p|}{|p_{n} - p|^{\alpha'}} = \lambda,
\newline
$$
then $\alpha'$ is the order of convergence, and $\lambda$ is the asymptotic error constant. Since our limiting value is zero (notice the independence from $\alpha'$), we can write this as $\newline \newline$
$$
\lim_{n \to \infty}\frac{|p_{n + 1}(\alpha)|}{|p_{n}(\alpha)|^{\alpha'}} = \lambda.
\newline
$$
Plugging in what $p_{n}(\alpha)$ is, we find $\newline \newline$
$$
\lim_{n \to \infty}\frac{10^{-\alpha^{n + 1}}}{\Big(10^{-\alpha^{n}}\Big)^{\alpha'}} = \lambda.
\newline
$$
This suggests we want to select $\alpha'$ to satisfy $-\alpha^{n + 1} = -\alpha^{n} \cdot \alpha'$, which gives us $\alpha = \alpha'$ and $\lambda = 1$. Therefore, what we've shown, is that, _in general_, $p_{n}(\alpha) = 10^{-\alpha^{n}}$ has order of convergence $\alpha$ for all $\alpha > 1$.

# Problem 3.3.8(a)

Use Algorithm 3.2 to construct the interpolating polynomial of degree four for the unequally spaced points given the following table:

| $x$ | $f(x)$ |
| --- | --- |
| $0.0$ | $-6.00000$ |
| $0.1$ | $-5.89483$ |
| $0.3$ | $-5.65014$ |
| $0.6$ | $-5.17788$ |
| $1.0$ | $-4.28172$ |

## Solution:

We need to construct the entries of the table

| $i$ | $x_{i}$ | $f[x_{i}]$ | $f[x_{i - 1}, x_{i}]$ | $f[x_{i - 2}, x_{i - 1}, x_{i}]$ | $f[x_{i - 3}, \dots, x_{i}]$ | $f[x_{i - 4}, \dots, x_{i}]$ |
| --- | --- | --- | --- | --- | --- | --- |
| $0$ | $0.0$ | $-6.00000$ |  | |  |  |
|  |  |  | $f[x_{0}, x_{1}]$ |  |  |  |
| $1$ | $0.1$ | $-5.89483$ |  | $f[x_{0}, x_{1}, x_{2}]$ |  |  |
|  |  |  | $f[x_{1}, x_{2}]$ |  | $f[x_{0}, x_{1}, x_{2}, x_{3}]$ |  |
| $2$ | $0.3$ | $-5.65014$ |  | $f[x_{1}, x_{2}, x_{3}]$ |  | $f[x_{0}, x_{1}, x_{2}, x_{3}, x_{4}]$ |
|  |  |  | $f[x_{2}, x_{3}]$ |  | $f[x_{1}, x_{2}, x_{3}, x_{4}]$ |  |
| $3$ | $0.6$ | $-5.17788$ |  | $f[x_{2}, x_{3}, x_{4}]$ |  |  |
|  |  |  | $f[x_{3}, x_{4}]$ |  |  |  |
| $4$ | $1.0$ | $-4.28172$ |  |  |  |  |

I will demonstrate calculation of all of the entries on the upper diagonal by hand in the interest of time and to suppress hand cramping; the other entries are computed similarly. If you'd like to check these numbers yourself, you can. I claim that the result of this computation gives

| $i$ | $x_{i}$ | $f[x_{i}]$ | $f[x_{i - 1}, x_{i}]$ | $f[x_{i - 2}, x_{i - 1}, x_{i}]$ | $f[x_{i - 3}, \dots, x_{i}]$ | $f[x_{i - 4}, \dots, x_{i}]$ |
| --- | --- | --- | --- | --- | --- | --- |
| $0$ | $0.0$ | $-6.00000$ |  | |  |  |
|  |  |  | $1.0517$ |  |  |  |
| $1$ | $0.1$ | $-5.89483$ |  | $0.5727$ |  |  |
|  |  |  | $1.2235$ |  | $0.2147$ |  |
| $2$ | $0.3$ | $-5.65014$ |  | $0.7015$ |  | $0.0633$ |
|  |  |  | $1.5742$ |  | $0.278$ |  |
| $3$ | $0.6$ | $-5.17788$ |  | $0.9517$ |  |  |
|  |  |  | $2.2404$ |  |  |  |
| $4$ | $1.0$ | $-4.28172$ |  |  |  |  |

To compute the upper diagonal, we did $\newline \newline$
$$
\begin{align*}
    f[x_{0}, x_{1}] = \frac{-5.89483 - (-6.00000)}{0.1 - 0} &= 1.0517 \\
    f[x_{0}, x_{1}, x_{2}] = \frac{1.2235 - 1.0517}{0.3 - 0} &= 0.5727 \\
    f[x_{0}, x_{1}, x_{2}, x_{3}] = \frac{0.7015 - 0.5727}{0.6 - 0} &= 0.2147 \\
    f[x_{0}, x_{1}, x_{2}, x_{3}, x_{4}] = \frac{0.278 - 0.2147}{1.0 - 0} &= 0.0633
\end{align*}
\newline
$$
Again, the other entries are computed in a similar manner. Now, to convert this table into a polynomial that interpolates the data, we read off the upper diagonal and construct the polynomial by $\newline \newline$
$$
\begin{align*}
P_{4}(x) &= f[x_{0}] \\
           &\;\;+ f[x_{0}, x_{1}](x - x_{0}) \\
           &\;\;+ f[x_{0}, x_{1}, x_{2}](x - x_{0})(x - x_{1}) \\
           &\;\;+ f[x_{0}, x_{1}, x_{2}, x_{3}](x - x_{0})(x - x_{1})(x - x_{2}) \\
           &\;\;+ f[x_{0}, x_{1}, x_{2}, x_{3}, x_{4}](x - x_{0})(x - x_{1})(x - x_{2})(x - x_{3}) \\
           &= -6.00000 \\
           &\;\;+ 1.0517(x - 0.0) \\
           &\;\;+ 0.5727(x - 0.0)(x - 0.1) \\
           &\;\;+ 0.2147(x - 0.0)(x - 0.1)(x - 0.3) \\
           &\;\;+ 0.063(x - 0.0)(x - 0.1)(x - 0.3)(x - 0.6) \\
           &= 0.0633x^{4} + 0.1514x^{3} + 0.5035x^{2} + 0.999766x - 6.
\end{align*}
\newline
$$

```julia
using Plots

data = [(0., -6.), (0.1, -5.89483), (0.3, -5.65014), (0.6, -5.17788), (1., -4.28172)]
P(x) = 0.0633*x^4 + 0.1514*x^3 + 0.5035*x^2 + 0.999766*x - 6.
xs = -0.5:0.01:1.5
plot(xs, P.(xs), label = "Interpolant", title = "Interpolation with Divided Differences", legend = :topleft)
scatter!(data, label = "Data")
```

# Problem 3.3.22 (Theoretical) and Discussion Question 2

Show that $\newline \newline$
$$
f[x_{0}, x_{1}, \dots, x_{n}, x] = \frac{f^{(n + 1)}(\xi(x))}{(n + 1)!}
\newline
$$
for some $\xi(x)$.

In light of the solution to the theoretical exercise, is it easier to add a new data pair using divided-difference methods or the Lagrange polynomial in order to obtain a higher-degree polynomial?

## Solution:

To show that $\newline \newline$
$$
f[x_{0}, x_{1}, \dots, x_{n}, x] = \frac{f^{(n + 1)}(\xi(x))}{(n + 1)!}
\newline
$$
for some $\xi(x)$, recall that the interpolating polynomial of degree $n$ for a dataset of $n + 1$ nodes is unique, and therefore the error formula for the Lagrange interpolant remains valid: $\newline \newline$
$$
f(x) = P(x) + \frac{f^{(n + 1)}(\xi(x))}{(n + 1)!}(x - x_{0})(x - x_{1})\dots(x - x_{n}).
\newline
$$
If we write out what the interpolating polynomial of degree $n + 1$ is on the nodes $x_{0}, x_{1}, \dots, x_{n}, x$, we find $\newline \newline$
$$
f(x) = P_{n + 1}(x) = P_{n}(x) + f[x_{0}, x_{1}, \dots, x_{n}, x](x - x_{0})\dots(x - x_{n})
\newline
$$
from the setup of the divided difference method. Given these two equations, it is clear then that $\newline \newline$
$$
f[x_{0}, x_{1}, \dots, x_{n}, x] = \frac{f^{(n + 1)}(\xi(x))}{(n + 1)!}
\newline
$$
for some $\xi(x)$.

Now, is it easier to use divided differences or Lagrange to assimilate new data into the interpolant?

<details>
  <summary>Answer (Click to Expand)</summary>
  It is easier to use divided differences. In light of the last solution, if you add a node on the ends of your original set (i.e., $x < x_{0}$ or $x > x_{n}$), then you can do exactly what was mentioned above. If you have $x$ in the middle of your old dataset, you will need to recompute a bit from your divided difference table, but it will not be so much effort (and the work you need to do does not scale when you add one more node in a small dataset versus adding one node in a large dataset). In contrast, if you wanted to add a node and use the Lagrange interpolant, you would need to recompute every single Lagrange polynomial and then add them together with the value of the function at the nodes. In other words, divided differences allows you to keep some old work, whereas Lagrange interpolation would make you throw it all away and start all over again. Thus, it's much easier (computationally speaking) to use divided differences to assimilate new data into your interpolating polynomial.
</details>

# Problem 3.4.1(c)

Use Theorem 3.9 or Algorithm 3.3 to construct an approximating polynomial for the following data.

| $x$ | $f(x)$ | $f'(x)$ |
| --- | --- | --- |
| $-0.5$ | $-0.0247500$ | $0.7510000$ |
| $-0.25$ | $0.3349375$ | $2.1890000$ |
| $0$ | $1.1010000$ | $4.0020000$ |

## Solution:

We construct the entries of the table for Algorithm 3.3:

| $z$ | $f(z)$ | $f[z_{i - 1}, z_{i}]$ | $f[z_{i - 2}, z_{i - 1}, z_{i}]$ | $f[z_{i - 3}, \dots, z_{i}]$ | $f[z_{i - 4}, \dots, z_{i}]$ | $f[z_{i - 5}, \dots, z_{i}]$ |
| --- | --- | --- | --- | --- | --- | --- |
| $\underline{-0.5}$ | $\underline{-0.0247500}$ | |  | |  | |
|  |  | $\underline{0.7510000}$ | |  |  | |
| $\underline{-0.5}$ | $\underline{-0.0247500}$ | | $2.751$ | |  | |
|  |  | $1.43875$ | | $1$ | | | |
| $\underline{-0.25}$ | $\underline{0.3349375}$ | | $3.001$ | | $0$ | |
|  |  | $\underline{2.1890000}$ | | $1$ | | $0$ |
| $\underline{-0.25}$ | $\underline{0.3349375}$ | | $3.501$ | | $0$  | |
|  |  | $3.06425$ | | $1$ |  | |
| $\underline{0}$ | $\underline{1.1010000}$ | | $3.751$ |  |  | |
|  |  | $\underline{4.0020000}$ | | |  | |
| $\underline{0}$ | $\underline{1.1010000}$ | |  |  |  | |

Then, the Hermite interpolating polynomial is $\newline \newline$
$$
H(x) = -0.02475 + 0.751(x + 0.5) + 2.751(x + 0.5)^{2} + 1(x + 0.5)^{2}(x + 0.25) + 0(x + 0.5)^{2}(x + 0.25)^{2} + 0(x + 0.5)^{2}(x + 0.25)^{2}(x - 0),
\newline
$$
which reduces to $\newline \newline$
$$
\begin{align*}
    H(x) &= -0.02475 + 0.751(x + 0.5) + 2.751(x + 0.5)^{2} + 1(x + 0.5)^{2}(x + 0.25) \\
    &= x^{3} + 4.001x^{2} + 4.002x + 1.101.
\end{align*}
\newline
$$

```julia
data = [(-0.5, -0.02475), (-0.25, 0.3349375), (0., 1.101)]
H(x) = x^3 + 4.001*x^2 + 4.002*x + 1.101
xs = -1.:0.01:0.5
plot(xs, H.(xs), label = "Interpolant", title = "Hermite Interpolation with Divided Differences", legend = :topleft)
scatter!(data, label = "Data")
```
